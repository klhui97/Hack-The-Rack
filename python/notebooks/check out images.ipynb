```python
import random
import pprint
from collections import Counter

import pandas as pd
from tqdm import tqdm
```

```python
df = pd.read_csv('tagged_results.csv')
```

```python
selected = df[['attributes', 'filename']]
```

```python
not_null = df[~df['attributes'].isnull()]
```

```python
not_null['parsed_attributes'] = not_null.dropna(axis=1)['attributes'].apply(lambda x: eval(str(x)))
```

```python
not_null['no_of_keys'] = not_null.dropna(axis=1)['parsed_attributes'].apply(lambda x: len(list(set(x.keys()))))
```

```python
not_null.dropna(axis=1).sort_values(by=['no_of_keys'], ascending=False)
```

```python
all_attributes = [list(not_null.iloc[i]['parsed_attributes'].keys()) for i in tqdm(range(len(not_null)))]
```

```python
flat_attributes = [item for sublist in all_attributes for item in sublist]
len(flat_attributes)
```

```python
counter = Counter(flat_attributes)
```

```python
attributes_df = pd.DataFrame([dict(counter)]).transpose()
```

```python
attributes_df.sort_values(by=[0], ascending=False) #Age Group Colo
```
